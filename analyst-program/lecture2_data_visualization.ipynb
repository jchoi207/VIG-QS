#### Numpy
- Libraries like Numpy provide improvements that make coding a lot more efficient 
- Useful functions for linear algebra, fourier transforms and matrices

```python
import numpy as np
```

#### Arrays
- Arrays come in various sizes and are the staple of Numpy
- Arrays are similar to lists; however, numpy stores array information continuously, making it faster when accessing.

```python
# 0 D array -> scalar
np.array(1)

# 1 D array -> vector which is similar to the list data structure from vanilla python
np.array([1,2,3,4])

# 2 D array -> matrix, which is like a list of lists 
my_array =  np.array([[1,2,3], [4,5,6]])
```

```python
print(my_array)
```

```python
nrow, ncol = my_array.shape ## returns the shape of my_array by row, and then column
print(nrow)
```

#### Indexing
- indexing is done in a similar fashion to Python; however, if you have an ndim array where ndim > 1, then you have to multi-index

```python
# [row, column]
# let's say we have some matrix
# [[1 2 3]
#  [4 5 6]]
# and we want to get access to number 6, 
# number 6, is in the 2nd row and the third column 

my_array[1, 2]
```

#### Matplot

- matplotlib is all about data visualization
- matplot helps us plot and visualize our data using graphs, histograms, bar charts, curves, etc...

```python
import matplotlib.pyplot as plt

x = [0,1,2,3,4]
y = [0,1,2,3,4]

x = np.array(x)
y = np.array(y)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('This is my first plot in matplot')
```

#### Plotting a scatter plot for the normal distribution
- Normal distribution is a bell curve that is symmetrical about the mean, at infinity, it approaches zero 
- Standard normal curve has a mean of 0, and a standard deviation of 1

```python
import matplotlib.pyplot as plt

l = []
for i in range(1000000):
    l.append(np.random.beta(100, 0.5))


plt.hist(l, bins=100, zorder=100, )
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('This is my first plot in matplot')
```

```python
import matplotlib.pyplot as plt

l = []
for i in range(100000):
    l.append(np.random.normal(2, 0.05))


plt.hist(l, bins=100, zorder=100, )
plt.xlim([0, 4])

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('This is my first plot in matplot')
```

#### Plotting Functions
- We can also plot functions using matplotlib instead of just plotting data points

```python
x = np.linspace(0, 100, 101)
y = np.sin(x)
print(x)
print(y)

plt.plot(x,y)
plt.title("Sine graph from 0, 100")
plt.xlabel('x')
plt.ylabel('y')
```

```python
x = np.linspace(0, 4* np.pi, 100)
y = np.sin(x)
print(x)
print(y)

plt.plot(x,y, c='g', linestyle='-.')
plt.title("Sine graph from 0, 100")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```
