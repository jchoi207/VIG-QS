#### Indexing Lists
- Indexing = Accessing
- An index is like the location of an entry within a list

````python
my_list = [0,1,2,3,4,5]

# print element at index 1
print(my_list[1])
print(my_list[1] == 1)

# printing the last element of a list
print(my_list[-1])
print(my_list[-1] == "apple")
````

### Functions
- Functions are a way to encapsulate code that you want to reuse.
- We can input data into a function, and get data out of a function

````python
## let's define a function that reads a list as input, and prints the list's elements

def my_list_reader(input_list):
    # anything that is tabbed is within the function
    # i want this function to print the elements of a list
    for element in input_list:
        print(element)
        
````

````python
## calling functions:

list_1 = [1,2,3]
list_2 = ['I', 'love', ' python']
list_3 = ['hello', 'True']

my_list_reader(list_1)
my_list_reader(list_2)
my_list_reader(list_3)
````

````python
## what if we want our function to output a variable
## example, let's take the sum of the elements in a list

def my_summation_list(input_list):
    ## takes in a list called input_list which only contains number
    ## returns the sum of the variables in input_list
    ## >> [10, 11, 13]
    ## >> 34
    return_sum = 0
    for element in input_list:
        return_sum = element + return_sum
    print(f"this is the output from within the function: {return_sum}")
    return return_sum

    #    = 10 + 0 = 10 
    #     = 11 + 10  =21 
    #     = 13 + 21 = 34 

x = my_summation_list([10,11,13])

print(x + 1)
````

#### Matplot
- matplotlib helps us plot and visualize our data using graphs, histograms, bar charts, curves, etc...

````python
import matplotlib.pyplot as plt

x = [0,1,2,3,4]
y = [0,1,2,3,4]

x = np.array(x)
y = np.array(y)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('This is my first plot in matplot')
````

#### Plotting a sampled normal distribution
- Normal distribution is a bell curve that is symmetrical about the mean, at infinity, it approaches zero 
- Standard normal curve has a mean of 0, and a standard deviation of 1

````python
import matplotlib.pyplot as plt

l = []
for i in range(1000000):
    l.append(np.random.normal(100, 0.5))


plt.hist(l, bins=100, zorder=100)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('This is my first plot in matplot')
````

#### Plotting Functions
- We can also plot functions using matplotlib instead of just plotting data points

````python
## let's use the np.linspace() function to generate x points

x = np.linspace(0, 100, 101)
y = np.sin(x)
print(x)
print(y)

plt.plot(x,y)
plt.title("Sine graph from 0, 100")
plt.xlabel('x')
plt.ylabel('y')
````

````python
x = np.linspace(0, 4* np.pi, 100)
y = np.sin(x)
print(x)
print(y)

plt.plot(x, y, c='g', linestyle='-.')
plt.title("Sine graph from 0, 100")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
````

#### Plotting Time Series Data using yfinance
- So far, we've been plotting functions and sampling from distributions, but what if we want to plot real data?
- We can use the yfinance library to get stock data from Yahoo Finance and then plot it using matplotlib

````python
import yfinance as yf
````

````python
## let's plot the YTD (year to date 2024-01-01 to today) at a daily frequency for AAPL
## Microsoft -> MSFT 
## TESLA -> TSLA

my_ticker = yf.Ticker('AAPL')
historical_data = my_ticker.history(start='2024-01-01', end='2024-08-08', interval = '1d')['Close']
print(historical_data)
````

````python
plt.title('AAPL share price in USD YTD')
plt.xlabel('time')
plt.ylabel('USD')
plt.grid()
plt.plot(historical_data.index, historical_data.values, c='r')
````

#### Comparing the performance of the magnificent 7
- Magnificent 7: Apple, Amazon, Google, Meta, Microsoft, Netflix, Tesla

````python
## in which stock did 100$ grow the most
## let's assume these stocks can be bought in non-integer amounts i.e. owning 0.33 shares of TSLA is allowed

tickers = [
    'AAPL',
    'AMZN',
    'GOOG',
    'META',
    'MSFT',
    'NFLX',
    'TSLA',
]
print(tickers)

for ticker in tickers:
    print(ticker)
    my_ticker = yf.Ticker(ticker)
    historical_data = my_ticker.history(start='2024-01-01', end='2024-08-08', interval = '1d')['Close']
    historical_data /= historical_data.values[0] 
    historical_data *= 100
    plt.plot(historical_data.index, historical_data.values, label=ticker)

plt.legend()
plt.title('AAPL share price in USD YTD')
plt.xlabel('time')
plt.ylabel('USD')
plt.grid()
````
